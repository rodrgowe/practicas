```python
import os
import shutil
import time
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
ruta = r'W:\2023\FINANZAS\INFORMACION PERIODICA'

def seleccionar_rutas():
    norma = entrada_norma.get()
    años = entrada_años.get()
    trimestre = entrada_trimestre.get()
    if norma == '1':
        a = os.listdir(ruta)
        rutas = []
        for i in os.listdir(ruta):
            if i.startswith((años)):
                b = os.path.join(ruta,i)
                rutas.append(b)
        rutas_a = []
        for x in rutas:
            subcarpetas = os.listdir(x)
            for i in subcarpetas:
                if i[:2] != '00':
                    b = os.path.join(x,i)
                    rutas_a.append(b)
    if trimestre == '1':
        trimestres = 'I','V'
    elif trimestre == '0':
        trimestres = 'V'
    rutas_a = []
    for i in a:
        if i.endswith((trimestres)) and i.startswith((años)):
            b = os.path.join(ruta,i)
            rutas_a.append(b)
    return rutas_a

def copiar_formatos():
    # Obtener los valores ingresados en la interfaz
    ruta_destino = entrada_ruta_destino.get()
    fecha_maxima = entrada_fecha_maxima.get()
    rutas_a = rutas_a
    formatos = entrada_formatos.get().split(',')  # Separa los formatos por coma
    
    # Parsear las fechas ingresadas
    try:
        fecha_minima = time.mktime(time.strptime(fecha_minima, "%d/%m/%Y"))
        fecha_maxima = time.mktime(time.strptime(fecha_maxima, "%d/%m/%Y"))
    except ValueError:
        resultado.config(text="Fechas inválidas. Asegúrate de usar el formato DD/MM/AAAA.")
        return

    # Tu código de copia de archivos aquí ...
    for rut in rutas_a:
        ruta_origen = rut
        ruta_origen = ruta_origen.replace('\\', '\\\\')
        ruta_desd = os.path.join(ruta_destino, os.path.basename(rut))
        # Obtener la lista de carpetas de empresas
        carpetas_empresas = [nombre for nombre in os.listdir(ruta_origen) if os.path.isdir(os.path.join(ruta_origen, nombre))]

        # Ingreso de la lista de formatos separados por comas


        # Recorrer cada carpeta de empresa
        for carpeta_empresa in carpetas_empresas:
            ruta_empresa = os.path.join(ruta_origen, carpeta_empresa)

            archivos_por_formato = {}  # Diccionario para almacenar los archivos por formato

            # Obtener la lista de archivos que cumplen con los formatos especificados y el rango de fechas
            for formato in formatos:
                archivos = [
                    nombre for nombre in os.listdir(ruta_empresa)
                    if nombre.startswith(formato) and fecha_minima <= os.path.getmtime(os.path.join(ruta_empresa, nombre)) <= fecha_maxima
                ]

                if archivos:
                    archivos_por_formato[formato] = archivos

            # Verificar si se encontraron archivos
            if archivos_por_formato:

                # Crear la carpeta de la empresa en la ruta de destino
                ruta_destino_empresa = os.path.join(ruta_desd, carpeta_empresa)
                os.makedirs(ruta_destino_empresa, exist_ok=True)

                # Copiar los archivos más recientes de cada formato a la carpeta de destino
                for formato, archivos in archivos_por_formato.items():
                    try:
                            
                        archivo_mas_reciente = max(archivos, key=lambda x: os.path.getmtime(os.path.join(ruta_empresa, x)))
                        ruta_origen_archivo = os.path.join(ruta_empresa, archivo_mas_reciente)
                        ruta_destino_archivo = os.path.join(ruta_destino_empresa, archivo_mas_reciente)
                        shutil.copy2(ruta_origen_archivo, ruta_destino_archivo)
                    except:
                        continue

# Función para seleccionar la ruta de destino
def seleccionar_ruta_destino():
    ruta = filedialog.askdirectory()
    entrada_ruta_destino.delete(0, tk.END)
    entrada_ruta_destino.insert(0, ruta)

# Crear la ventana principal
ventana = tk.Tk()
ventana.title("Copiar Archivos")

# Crear y configurar etiquetas y entradas para cada campo
etiqueta_norma = ttk.Label(ventana, text="Norma (0 = 096, 1 = 043)")
entrada_norma = ttk.Entry(ventana)
etiqueta_años = ttk.Label(ventana, text="Años")
entrada_años = ttk.Entry(ventana)
etiqueta_trimestre = ttk.Label(ventana, text= "0 =IV trimestre, 1 = I,II,III,IV")
entrada_trimestre = ttk.Entry(ventana)
etiqueta_ruta_destino = ttk.Label(ventana, text="Ruta de destino:")
entrada_ruta_destino = ttk.Entry(ventana)
etiqueta_fecha_maxima = ttk.Label(ventana, text="Fecha máxima (DD/MM/AAAA):")
entrada_fecha_maxima = ttk.Entry(ventana)
etiqueta_formatos = ttk.Label(ventana, text="Formatos (separados por coma):")
entrada_formatos = ttk.Entry(ventana)

# Botón para iniciar la copia
boton_copiar = ttk.Button(ventana, text="Copiar Archivos", command=copiar_formatos)

# Ubicar los elementos en la ventana usando grid
etiqueta_norma.grid(row=0, column=0, padx=5, pady=5)
entrada_norma.grid(row=0, column=1, padx=5, pady=5)
etiqueta_años.grid(row=1, column=0, padx=5, pady=5)
entrada_años.grid(row=1, column=1, padx=5, pady=5)
etiqueta_trimestre.grid(row=2, column=0, padx=5, pady=5)
entrada_trimestre.grid(row=2, column=1, padx=5, pady=5)
etiqueta_ruta_destino.grid(row=3, column=0, padx=5, pady=5)
entrada_ruta_destino.grid(row=3, column=1, padx=5, pady=5)
etiqueta_fecha_maxima.grid(row=5, column=0, padx=5, pady=5)
entrada_fecha_maxima.grid(row=5, column=1, padx=5, pady=5)
etiqueta_formatos.grid(row=6, column=0, padx=5, pady=5)
entrada_formatos.grid(row=6, column=1, padx=5, pady=5)
boton_copiar.grid(row=7, column=0, columnspan=2, pady=10)


# Botones para seleccionar rutas

boton_seleccionar_ruta_destino = ttk.Button(ventana, text="Seleccionar", command=seleccionar_ruta_destino)
boton_seleccionar_ruta_destino.grid(row=3, column=2, padx=5, pady=5)

# Etiqueta para mostrar el resultado de la operación
resultado = ttk.Label(ventana, text="", foreground="green")
resultado.grid(row=6, column=0, columnspan=2, pady=10)

# Ejecutar la ventana
ventana.mainloop()
```
