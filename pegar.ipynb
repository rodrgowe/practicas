```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook
import os
```

```python
ruta = r'C:\Users\User TP3\Desktop\moneda extranjera'
data = []
for archivo in os.listdir(ruta):
    if archivo.lower().endswith('.xlsx'):
        ruta_entera = os.path.join(ruta,archivo)
        wb = load_workbook(ruta_entera)
        sheet = wb.active
        val_celda = sheet['B4'].value
        fecha = str(val_celda)[-10:]
        df = pd.read_excel(ruta_entera, skiprows=6,skipfooter=35)
        df = df.drop(df.columns[0], axis=1)
        df = df.iloc[7:]
        df['fecha'] = fecha
        data.append(df)
```

```python
data = pd.concat(data, ignore_index=True)
```
