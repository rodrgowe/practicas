```python
columnas = ['Mes', 'Departamento', 'Provincia', 'Distrito', 'Centro Poblado',
       'Código de Centro Poblado', 'Segmento', 'Tecnología',
       'Velocidad Contratada - Bajada (Mbps)',
       'Velocidad Contratada -Subida (Mbps)',
       'Velocidad  Garantizada - Bajada (Mbps)',
       'Velocidad  Garantizada - Subida (Mbps)', 'Conexiones en servicio','Año','Trimestre', 'Formato', 'Empresa', 'Fecha']
enteros = ['Velocidad Contratada - Bajada (Mbps)',
       'Velocidad Contratada -Subida (Mbps)',
       'Velocidad  Garantizada - Bajada (Mbps)',
       'Velocidad  Garantizada - Subida (Mbps)', 'Conexiones en servicio']
```

```python
datos = []
for ruta, carpetas, archivos in os.walk(wd):
    for archivo in archivos:
        if archivo.startswith(formato_str) and archivo.lower().endswith(".xlsx"):
            try:
                
                ruta_completa = os.path.join(ruta, archivo)
                # Obtener datos de las celdas
                wb = load_workbook(ruta_completa)
                sheet = wb.active
                year = sheet['C16'].value
                trimestre = sheet['E16'].value
                df1 = pd.read_excel(ruta_completa,skiprows=20)
                df1['Año'] = year
                df1['Trimestre'] = trimestre
                # Obtener la primera y segunda parte del nombre del formato
                nombre_formato = os.path.splitext(archivo)[0]  # Obtener el nombre del archivo sin extensión
                partes = nombre_formato.split(" - ")
                partes1 = nombre_formato.split(" - ")[1].split(" ")
                partes_nombre = nombre_formato.split(' - ')
                df1['Formato'] = partes[0]
                df1['Empresa'] = " ".join(partes1[0:-1])
                df1['Fecha'] = partes1[-1]
                df1 = df1.drop(df1.columns[0], axis=1)  # Eliminar la primera columna
                datos.append(df1)
            except: 
                print(f'error con el {archivo}')
```

```python
acf["Dep+Prov"] = acf["Departamento"] + " " + acf["Provincia"]

# se crea la columna que concatena los valores de DEPARTAMENTO, PROVINCIA y DISTRITO
acf["Dep+Prov+Dist"] = acf["Dep+Prov"] + " " + acf["Distrito"]

# se crea la columna que concatena los valores de DEPARTAMENTO, PROVINCIA, DISTRITO y CCPP
acf["Dep+Prov+Dist+CCPP"] = acf["Dep+Prov+Dist"] + " " + acf["Centro Poblado"]
```

```python
#en la base 
# se crea la columna que concatena los valores de DEPARTAMENTO y PROVINCIA
# se crea la columna que concatena los valores de DEPARTAMENTO y PROVINCIA
ccpp_base["Dep+Prov"] = ccpp_base["Departamento"] + " " + ccpp_base["Provincia"]

# se crea la columna que concatena los valores de DEPARTAMENTO, PROVINCIA y DISTRITO
ccpp_base["Dep+Prov+Dist"] = ccpp_base["Dep+Prov"] + " " + ccpp_base["Distrito"]

# se crea la columna que concatena los valores de DEPARTAMENTO, PROVINCIA, DISTRITO y CCPP
ccpp_base["Dep+Prov+Dist+CCPP"] = ccpp_base["Dep+Prov+Dist"] + " " + ccpp_base["Centro Poblado"]

##VALIDAR DATOS 

for column in acf:
    if column in valores.columns:
        acf[column + '_verif_BD'] = acf[column].isin(valores[column]).map({True: "Valor Permisible", False:"Valor no permisible"})
    elif column in ccpp_base.columns:
        acf[column + '_verif_CCPP'] = acf[column].isin(ccpp_base[column]).map({True: "Valor Permisible", False:"Valor no permisible"})
    else:
        if column in enteros:
            acf[column + '_verif_ent'] = acf[column].apply(lambda x: "Valor no permisible" if pd.notnull(x) and not isinstance(x, (int, float)) else "Valor en Blanco" if pd.isnull(x) else "Valor Permisible")
        else:
            acf[column + '_verif_NA'] = acf[column].isna().map({True: "Valor en Blanco", False:"Valor Permisible"})
```

```python

# Definir la función para aplicar la condición
def inalambrica(row):
    if row['Tecnología'].upper() in ['ADSL','ADSL2','ADSL2+']:
        return 'xDSL'
    elif row['Tecnología'].upper() in ['DOCSIS','EOC']:
        return 'Cable Módem'
    elif row['Tecnología'].upper() in ['GPON', '10-GPON', 'AON', 'EPTP']:
        return 'FTTH'
    elif row['Tecnología'].startswith('Otros Inalámbricos'):
        return 'Otros Inalámbricos'
    elif row['Tecnología'].startswith('Otros Alámbricos'):
        return 'Otros Alámbricos'
    elif row['Tecnología'].lower() in ['ftth']:
        return 'FTTH'
    else:
        return row['Tecnología']  # Otra opción si no se cumple ninguna condición
```

```python
# Aplicar la función a lo largo de las filas y crear la nueva columna
acf['Tecno_modi'] = acf.apply(inalambrica, axis=1)
```
